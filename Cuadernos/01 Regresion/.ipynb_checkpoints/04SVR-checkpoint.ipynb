```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importar el Data Set
dataSet = pd.read_csv("Position_Salaries.csv")
dataSet
```

```python
X = dataSet.iloc[:, 1:2].values
y = dataSet.iloc[:, 2].values
```

```python
print("type(X): ", type(X), ", Filas y Columnas X: ", X.shape, ", Filas de X: ", X.shape[0])
print("type(y): ", type(y), ", Filas y Columnas y: ", y.shape, ", Filas de y: ", y.shape[0])
```

```python
dataSet
```

```python
#Visualizar Conjunto de Datos Originales
from matplotlib.ticker import FuncFormatter, ScalarFormatter
plt.scatter(X, y, color = "red")
plt.title("Salario en funcion de Nivel")
plt.xlabel("Nivel")
plt.ylabel("Salario")
plt.show()
```

```python
#Escalar variables
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y.reshape(-1,1))
```

```python
print("type(X): ", type(X), ", Filas y Columnas X: ", X.shape, ", Filas de X: ", X.shape[0])
print("type(y): ", type(y), ", Filas y Columnas y: ", y.shape, ", Filas de y: ", y.shape[0])
```

```python
#Visualizar Conjunto de Datos Escalados
from matplotlib.ticker import FuncFormatter, ScalarFormatter
plt.scatter(X, y, color = "red")
plt.title("Datos Escalados")
plt.xlabel("Nivel")
plt.ylabel("Salario")

plt.show()
```

```python
#Ajustar la regresin con el Data Set
from sklearn.svm import SVR
regression = SVR(kernel = "rbf", gamma = "auto")
regression.fit(X,y.ravel())
```

```python
#Mostrar Hiper Parametros
regression.get_params()
```

```python
#Visualizar los resultados del SVR
plt.scatter(X, y, color = "red")
plt.plot(X, regression.predict(X), color = "blue")
plt.title("Salario en funcion del Nivel")
plt.xlabel("Nivel")
plt.ylabel("Salario")
plt.show()
```

```python
print("X mean:", sc_X.mean_)
print("X std:", sc_X.scale_)
```

```python
#Prediccion de nuestros modelos SVR
valorAPredecir = np.array([[6.5]])
valorEscalado = sc_X.transform(valorAPredecir)

y_pred = regression.predict(valorEscalado)
y_pred
```

```python
#Reentrena el SVR explcitamente con gamma='auto' y predice
X_raw = dataSet.iloc[:, 1:2].values
y_raw = dataSet.iloc[:, 2].values

# Copia los X originales del CSV para evitar usar versiones modificadas o escaladas previamente.
Xr = X_raw.copy()
# Copia y y lo convierte en columna (2D) porque StandardScaler exige 2D.
yr = y_raw.copy().reshape(-1,1)

# Crea dos scalers: uno para X y otro para y (no pueden compartir scaler).
scX = StandardScaler(); scY = StandardScaler()
# Ajusta scX usando Xr y luego transforma Xr  ahora Xr_s est escalado.
Xr_s = scX.fit_transform(Xr)
# Ajusta scY a yr, transforma yr y luego lo aplasta a 1D (ravel) porque SVR exige y 1D.
Yr_s = scY.fit_transform(yr).ravel()

svr = SVR(kernel="rbf", gamma="auto")   # fuerza gamma='auto'
svr.fit(Xr_s, Yr_s)

# predecir para 6.5 con el reshape correcto
valor = np.array([[6.5]])# Convierte 6.5 en un array 2D para que el scaler pueda procesarlo.
valor_s = scX.transform(valor)# Escala 6.5 usando el mismo scaler con el que se entren el modelo.
y_pred_scaled = svr.predict(valor_s)
y_pred_original = scY.inverse_transform(y_pred_scaled.reshape(-1,1))

print("y_pred (escalado):", y_pred_scaled)
print("y_pred (Salario Equivalente):", y_pred_original.ravel())
```
